```
## Concatenate all previous Dataframes into one master Dataframe and drop any Na records that exist 

lyrics_final_df = pd.concat([df1,df2,df3,df4,df5,df6,df7],ignore_index=True)
lyrics_final_df = lyrics_final_df[['artist','title','lyrics','tokenized_lyrics']]
lyrics_final_df = lyrics_final_df.dropna()

lyrics_final_df.tail(50)

```

```
## Take a copy of the final dataframe and assign it to lyrics_exploded_df. Then explode the tokenized lyrics and drop any na's, having each word rowise will help
## with filtering only artist who meet the greater than 5000 overall lyric count

lyrics_exploded_df = lyrics_final_df.copy()
lyrics_exploded_df = lyrics_final_df.explode('tokenized_lyrics').dropna()
lyrics_exploded_df.head()
```

```
## Group by artist and count the total count of lyrics from the Artists body of work 
## save the results to test2.csv to keep intermediate results

testdf=lyrics_exploded_df[['artist','tokenized_lyrics']].groupby('artist').agg('count').sort_values(by ='tokenized_lyrics', ascending=False).reset_index()
testdf.to_csv('/work/lyrical_analysis/test2.csv')
```

```
# Filtering out Paradox and Riot after doing a deep dive on the data. We found that Genius API did not return only the lyrics for those bands, but 
# any bands that had the similar names, or part of the name in the band. So it was literally doing a search on the Band name that was passed
# As a result, we removed paradox and riot from our analysis  and set a filter on only analyzing bands that had greater than 5000 total lyric count

filter_artists = ['Paradox','Riot']

testdf3 = testdf[testdf["tokenized_lyrics"] > 4999].reset_index()
testdf3 = testdf3[~testdf3['artist'].isin(filter_artists)].reset_index()
testdf3.head()

```

```
## we applied a left join to the inital lyrics_exploded_df with the testdf3 dataframe, to include the tokenized_lyrics_x column in the dataframe,
## which is the total lyric count

testdf4 = testdf3.merge(lyrics_exploded_df, on='artist', how='left')
testdf4.head()
```

```
## A new column was added and set to have a value of 1, to help us find the truely unique words in the dataset per each artist 

testdf4['add_one'] = 1
testdf4.head()
```

```
# new_test_df groups by arist and tokenized_lyrics_y and sums the add_one columns, we then only filter for when add_one column is 1
# new_test_df results are then left joined with testdf3 to added the total count of unique words which is then passed to metal_unique_df 
# A new column is added, unique_ratio which calculates the unique word count divided by the total word count times a 100 , this will be used
# for analysis and visualization step 

new_test_df = (testdf4[['artist','tokenized_lyrics_x','tokenized_lyrics_y','add_one']]
    .groupby(['artist','tokenized_lyrics_y'])
    .sum('add_one')
    .reset_index())


new_test_df.head(100)


new_test_df = new_test_df[new_test_df['add_one'] < 2].reset_index()
new_test_df = (new_test_df[['artist','add_one']]
    .groupby(['artist'])
    .sum('add_one')
    .reset_index())

metal_unique_df = testdf3.merge(new_test_df, on='artist', how='left')
metal_unique_df['unique_ratio'] = (metal_unique_df['add_one'] / metal_unique_df['tokenized_lyrics']) * 100

metal_unique_df.sort_values(by = 'unique_ratio',ascending=False)
metal_unique_df.to_csv('/work/lyrical_analysis/metal_analysis.csv')
```
